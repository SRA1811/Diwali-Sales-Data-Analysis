# Exloratory Analysis

### Gender

```python
df.columns
```

```python
df['Gender'].value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.show()
```

```python
plt.figure(figsize=(10,5))
a=sns.countplot(data=df, x='Gender')

for bars in a.containers:
    a.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sales_gen
```

```python
a=sns.barplot(x='Gender', y='Amount', data=sales_gen)
for bars in a.containers:
    a.bar_label(bars)
```

## Age

```python
plt.figure(figsize=(10,5))
a=sns.countplot(x='Age Group', hue='Gender', data=df)
b=a.bar_label(a.containers[0])
c=a.bar_label(a.containers[1])
```

```python
Sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
Sales_age
```

```python
plt.figure(figsize=(10,5))
Sales_age=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
a=sns.barplot(data=Sales_age, x='Age Group', y='Amount')
for bars in a.containers:
    a.bar_label(bars)
```

## State

```python
sales_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sales_state
```

### Total number of orders from top 10 states.

```python
sns.set(rc={'figure.figsize':(20,10)})
orders_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
a=sns.barplot(data=orders_state,x='State', y='Orders')
for bars in a.containers:
    a.bar_label(bars)
```

### Total amount of sales from top 10 states.

```python
sns.set(rc={'figure.figsize':(20,10)})
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
a=sns.barplot(data=sales_state, x='State', y='Amount')
for bars in a.containers:
    a.bar_label(bars)
```

*From above visualizatione we can see that most of the sales amount and orders are from Uttar Pradesh, Maharashtra and Karnataka respectively.*

## Maritial Status

```python
sns.set(rc={'figure.figsize':(7,5)})
a=sns.countplot(x='Marital_Status', data=df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
m_s=df.groupby(['Marital_Status','Gender'], as_index=False)["Amount"].sum().sort_values(by='Amount', ascending=False)
sns.set(rc={'figure.figsize':(7,5)})
a=sns.barplot(data=m_s, x='Marital_Status', y='Amount', hue='Gender')
for bars in a.containers:
    a.bar_label(bars)
```

*From above visualizarion we can see that most of the buyers are married (Women) and have high purchasing power.*

## Occupation

```python
sns.set(rc={'figure.figsize':(20,10)})
a=sns.countplot(data=df, x='Occupation').sort_valuessns
for bars in a.containers:
    a.bar_label(bars)
```

```python
occ_amount=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
occ_amount
```

```python
a=sns.barplot(data=occ_amount, x='Occupation', y='Amount')
for bars in a.containers:
    a.bar_label(bars)
```

*From above graph we can see that most of the buyers are from IT, Healthcare and Aviation sector.*

## Product Category

```python
sns.set(rc={'figure.figsize':(20,10)})
a=sns.countplot(data=df,x='Product_Category')
for bars in a.containers:
    a.bar_label(bars)
plt.xticks(rotation=60)
plt.show()
```

*From above graph we can see that most of the sold produbcts are from Clothing & appearal, food and electronics.*

```python
pc_amt=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
a=sns.barplot(data=pc_amt,x='Product_Category',y='Amount')
plt.xticks(rotation=60)
plt.show()
```

```python
top_prod=df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
top_prod
```

```python
a=sns.barplot(data=top_prod, x='Product_ID', y='Orders')
for bars in a.containers:
    a.bar_label(bars)
```

*Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
